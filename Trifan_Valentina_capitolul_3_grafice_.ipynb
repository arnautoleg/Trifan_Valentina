```python
import matplotlib.pyplot as plt
import numpy as np
import math

def confidence_interval(count, total, confidence=0.95):
    p = count / total
    z = 1.96  # Z-score for 95% confidence level

    # Calculate standard error
    se = math.sqrt((p * (1 - p)) / total)

    # Calculate margin of error
    margin_error = z * se

    # Calculate lower and upper bounds
    lower_bound = p - margin_error
    upper_bound = p + margin_error

    return (lower_bound, upper_bound)

# Sample data (counts)
categories = ['Rural', 'Urban']
total = 1710  # Total count
counts = [47.8*total/100, 55.2*total/100]  # Sample counts for each category

# Calculate percentages and confidence intervals
percentages = [count / total * 100 for count in counts]
confidence_intervals = [confidence_interval(count, total) for count in counts]

# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, percentages, color='skyblue')

# Add percentages and confidence intervals as annotations
for i, bar in enumerate(bars):
    lower, upper = confidence_intervals[i]
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 4, f'{percentages[i]:.1f}%', fontsize=14)
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.2, bar.get_height() + 1, f'95%CI: [{lower * 100:.1f}, {upper * 100:.1f}]', fontsize=12)

# Add title and labels
plt.title('Bar Plot of Percentages with 95% Confidence Intervals')
plt.xlabel('Categoriile')
plt.ylabel('Frecventa relativa, %')

# Show plot
# Set y-axis limits
plt.ylim(0, 100)
plt.tight_layout()
plt.show()




```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories1 = ['RDD Nord', 'RDD Centru', 'RDD Sud']
categories2 = ['Altele', 'Nu lucreaza', 'Intelectual', 'Muncitor', 'Agricultor']
data = np.array([
    [10.4, 10.8, 2.2],
    [12.4, 8.8, 10.2],
    [21.2, 26.2, 28.2],
    [47.0, 31.9, 46.9],
    [9, 22.3, 12.5]
])

# Transpose the data
data_transposed = data.T

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))  # Width = 10 inches, Height = 6 inches
width = 0.3  # Width of each bar
gap_width = 0.2  # Gap width between groups of bars
x = np.arange(len(categories2))

for i, region_data in enumerate(data_transposed):
    ax.bar(x + width * i, region_data, width, label=categories1[i])

ax.set_ylabel('Frecventa relativa, %')
ax.set_ylim(0, 100)
ax.set_xlabel('Categorie')
ax.set_xticks(x + width * (len(categories1) - 1) / 2)
ax.set_xticklabels(categories2)
ax.legend()

# Add values on the bars
for i, region_data in enumerate(data_transposed):
    for j, value in enumerate(region_data):
        ax.text(x[j] + width * i, value + 1, str(value), ha='center')

plt.show()
```
